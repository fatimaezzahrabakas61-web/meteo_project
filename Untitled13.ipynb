```python
import pandas as pd

# 1. Charger le fichier
# =========================
def charger_fichier(path):
    return pd.read_excel(path)

# convertir au numerique
# =========================
def convertir_en_numerique(df):
    colonnes = [
        "Apports",
        "Evaporation",
        "Stock au de la C.A 1 er Sep"
    ] # Removed 'TOTAL' as it was causing a KeyError

    for col in colonnes:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    return df

df = charger_fichier("copy-hydraulique_touse.xlsx")
df = convertir_en_numerique(df)
```

```python
# =========================
# 2. Ajouter colonne Transfert eau (valeurs numériques)
# =========================
def ajouter_transfert_eau(df):
    def mapper(ca):
        if ca == "2022-2023":
            return 20
        elif ca == "2021-2022":
            return 40
        else:
            return 0

    df["Transfert_eau"] = df["C_A"].apply(mapper).astype(int)
    return df

df = ajouter_transfert_eau(df)
```

```python
# 3. Calcul Total corrigé
# =========================
def calcul_total_corrige(df):
    df["Apports"] = pd.to_numeric(df["Apports"], errors='coerce').fillna(0).astype(int) # Ensure 'Apports' is numeric
    df["Total_corrige"] = df["Apports"] + df["Transfert_eau"]
    return df

df = calcul_total_corrige(df)
```

```python
# 4. Calcul Stock final
# =========================
def calcul_stock_final(df):
    df["Stock_final"] = (
        df["Stock au de la C.A 1 er Sep"] + df["Total_corrige"]
        - (df["Evaporation"])
    ) # Removed df["TOTAL"] as it does not exist in the DataFrame
    return df

df = calcul_stock_final(df)
```

```python
# 5. Calcul Stock utile
# =========================
def calcul_stock_utile(df):
    df["Stock_utile"] = df["Stock_final"] - 61
    return df
df = calcul_stock_utile(df)
```

```python
def sauvegarder_fichier(df, path):
    df.to_excel(path, index=False)

# Call the function with the processed DataFrame 'df'
# (assuming 'df' has been correctly created and updated by previous cells)
# The function does not return a DataFrame, so remove the assignment to 'df'.
sauvegarder_fichier(df, "resultat_final.xlsx")
```
