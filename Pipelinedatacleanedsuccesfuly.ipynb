```python
import pandas as pd

# 1. Charger le fichier
df = pd.read_csv("dataset-noaa-Copy-toUse.csv", sep=',', skiprows=1)

# 2. Renommer colonnes
df.rename(columns={
    'Date': 'DATE',
    'TAVG (Degrees Fahrenheit)': 'TAVG',
    'TMAX (Degrees Fahrenheit)': 'TMAX',
    'TMIN (Degrees Fahrenheit)': 'TMIN',
    'PRCP (Inches)': 'PRCP',
    'SNOW (Inches)': 'SNOW',
    'SNWD (Inches)': 'SNWD',
}, inplace=True)

# -----------------------------
# 🔹 1. CONVERTIR EN NUMÉRIQUE
# -----------------------------
cols = ["TAVG", "TMAX", "TMIN", "PRCP"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# 👉 transforme les valeurs non numériques en NaN

# -----------------------------
# 🔹 2. GÉRER LES NaN
# -----------------------------
df[cols] = df[cols].interpolate(method="linear")

# 👉 remplace les valeurs manquantes par interpolation
# 👉 garde le maximum de données

# -----------------------------
# 🔹 3. APPLIQUER LES RÈGLES
# -----------------------------

# ✔️ Cohérence physique
df = df[df["TMAX"] >= df["TMIN"]]

# 👉 supprime lignes incohérentes (TMAX < TMIN)

# ✔️ Conversion unités
df["PRCP"] = df["PRCP"] * 25.4  # inches → mm

df["TAVG"] = (df["TAVG"] - 32) * 5/9
df["TMAX"] = (df["TMAX"] - 32) * 5/9
df["TMIN"] = (df["TMIN"] - 32) * 5/9

# 👉 convertit en unités standards (mm, °C)

# -----------------------------
# 🔹 4. FEATURE ENGINEERING
# -----------------------------
df["TEMP_RANGE"] = df["TMAX"] - df["TMIN"]

df["TAVG_7D"] = df["TAVG"].rolling(7).mean()
df["PRCP_7D"] = df["PRCP"].rolling(7).mean()

# 👉 crée nouvelles variables utiles

# -----------------------------
# 🔹 5. ARRONDIR
# -----------------------------
df = df.round(2)

# 👉 limite à 2 chiffres après virgule
df = df.drop(columns=['SNOW', 'SNWD'])

# -----------------------------
# 🔹 6. SAUVEGARDE
# -----------------------------
df.to_csv("cleaned_data.csv", index=False)

print("✅ Nettoyage terminé")
```
