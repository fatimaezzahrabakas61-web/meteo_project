```python
import pandas as pd
# 1. read file
def load_data(path):
    return pd.read_excel(path)

# 2. Convertir les colonnes en numérique
def convert_columns_to_numeric(df, cols):
    for col in cols:
        df[col] = pd.to_numeric(
            df[col].astype(str).str.replace(",", "."),
            errors="coerce"
        ).fillna(0)
    return df


# 3. Ajouter la colonne transfert_eau
def create_transfert_eau(df, source_col="C.A"):
    df["transfert_eau"] = df[source_col].apply(
        lambda x: 40 if x == "2021-2022" else 0
    )
    return df
# 4. Ajouter start_date et end_date
def add_date_columns(df, start_year=1937):
    """
    Ajoute les colonnes start_date et end_date
    """
    n = len(df)

    df["start_date"] = list(range(start_year, start_year + n))
    df["end_date"] = list(range(start_year + 1, start_year + n + 1))

    return df

# 5. Renommer les colonnes
def rename_ca_column(df):
    return df.rename(columns={"C.A": "years"})

def rename_months_columns(df):
    """
    Renomme les colonnes des mois en anglais standard
    """
    month_mapping = {
        "Jan": "Jan",
        "Fev": "Feb",
        "Mar": "Mar",
        "Avril": "Apr",
        "Mai": "May",
        "Juin": "Jun",
        "Juillet": "Jul",
        "Aout": "Aug",
        "Sep": "Sep",
        "Oct": "Oct",
        "Nov": "Nov",
        "Dec": "Dec"
    }

    df = df.rename(columns=month_mapping)
    return df


# 6. Sauvegarder
def save_data(df, path):
    df.to_csv(path, index=False)

# 7. Fonction principale (pipeline)
def preprocess_dataset(df, output_path):
    # Colonnes à convertir
    mois_total_cols = ["Jan", "Fev", "Mar", "Avril", "Mai", "Juin",
                   "Juillet", "Aout", "Sep", "Oct", "Nov", "Dec", "Total"]


    df = convert_columns_to_numeric(df, mois_total_cols)
    df = create_transfert_eau(df)
    df = add_date_columns(df)
    df = rename_ca_column(df)
    df = rename_months_columns(df)
    save_data(df, output_path)
    return df
```

```python
# ▶️ Exécution
if __name__ == "__main__":
   input_file = "Apports-copyToUse.csv"
   output_file = "dataset_Apports_clean.csv"

   df_input = pd.read_csv(input_file) # Read the CSV into a DataFrame
   preprocess_dataset(df_input, output_file)
```

```python
df_raw = pd.read_csv("Apports-copyToUse.csv")
df = preprocess_dataset(df_raw)
```

```python
df.to_excel("dataset_apports_clean.csv", index=False)
```
