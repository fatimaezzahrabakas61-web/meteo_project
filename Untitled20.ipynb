```python
import pandas as pd


# 1. Charger les données
def load_data(path):
    return pd.read_excel(path)
```

```python
def extract_volumes(df):
    volume_cols = [
        "Stock_au_de_C.A_1erSep",
        "Apports",
        "Evaporation",
        "Resti_total"
    ]

    # garder seulement les colonnes existantes
    volume_cols = [col for col in volume_cols if col in df.columns]

    return df[volume_cols]
```

```python
def extract_entrees_sorties(df):
    entrees_cols = ["Apports"]
    sorties_cols = ["Evaporation", "Resti_total"]

    entrees_cols = [col for col in entrees_cols if col in df.columns]
    sorties_cols = [col for col in sorties_cols if col in df.columns]

    df_entrees = df[entrees_cols]
    df_sorties = df[sorties_cols]

    return df_entrees, df_sorties
```

```python
def check_coherence(df):
    required_cols = [
        "Stock_au_de_C.A_1erSep",
        "Apports",
        "Evaporation",
        "Resti_total"
    ]

    # vérifier que toutes les colonnes existent
    for col in required_cols:
        if col not in df.columns:
            df[col] = 0

    # calcul du bilan
    df["bilan"] = (
        df["Stock_au_de_C.A_1erSep"]
        + df["Apports"]
        - df["Evaporation"]
        - df["Resti_total"]
    ).abs()



    return df
```

```python
def save_data(df, path):
    df.to_excel(path, index=False)

def pipeline(input_path, output_path):
  df = load_data(input_path)
  df = check_coherence(df)

  save_data(df, output_path)

    # extraction (optionnelle)
  df_volumes = extract_volumes(df)
  df_entrees, df_sorties = extract_entrees_sorties(df)

  return df, df_volumes, df_entrees, df_sorties
```

```python
# ▶️ Exécution
if __name__ == "__main__":
   input_file = "dataset_bilanhyd_clean (3).xlsx"
   output_file = "dataset_bilanhybri_clean.xlsx"

   pipeline(input_file, output_file)
```
