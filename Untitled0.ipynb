```python
import pandas as pd

# Charger
df = pd.read_csv("dataset-noaa-Copy-toUse.csv")

# Transformer MultiIndex en colonnes
df = df.reset_index()

# Renommer colonnes
df.columns = ['Date', 'TAVG', 'TMAX', 'TMIN', 'PRCP', 'SNOW', 'STATION']

df.head()
```

```python
df.columns = ['Date', 'TAVG', 'TMAX', 'TMIN', 'PRCP', 'SNOW', 'STATION']
```

```python
df.head()
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df['Date'] = pd.to_datetime(df.index)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method='ffill')
df = df.fillna(method='bfill')
```

```python
df = df[df['TMAX'] >= df['TMIN']]
```

```python
df['TAVG'] = (df['TAVG'] - 32) * 5/9
df['TMAX'] = (df['TMAX'] - 32) * 5/9
df['TMIN'] = (df['TMIN'] - 32) * 5/9
```

```python
df = df[(df['TMAX'] < 60)]
df = df[(df['TMIN'] > -10)]
df = df[(df['PRCP'] >= 0)]
```

```python
df['PRCP'] = df['PRCP'] * 25.4
```

```python
df['temp_amplitude'] = df['TMAX'] - df['TMIN']
```

```python
df['TAVG_MA7'] = df['TAVG'].rolling(7).mean()
df['PRCP_MA7'] = df['PRCP'].rolling(7).mean()
```

```python
df['dry_index'] = df['PRCP'] / (df['TAVG'] + 1)
```

sauvegarde des modification dans dataset

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
