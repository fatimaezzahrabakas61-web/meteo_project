```python
import pandas as pd
import matplotlib.pyplot as plt

def load_and_prepare_data(filepath):
    """
    Charger le dataset et créer year_start à partir de start_date
    """
    df = pd.read_csv(filepath)
    df["year_start"] = range(1972, 1972 + len(df))
    return df
```

```python
def plot_evaporation(df):
    """
    Evolution de l'évaporation par year_start
    """
    plt.figure()
    plt.plot(df["year_start"], df["evaporation"])
    plt.xlabel("Year")
    plt.ylabel("Evaporation")
    plt.title("Evolution of Evaporation Over Years")
    plt.show()
```

```python
def plot_water_balance(df):
    """
    Evolution du bilan hydrique par year_start
    """
    plt.figure()
    plt.plot(df["year_start"], df["water_balance"])
    plt.xlabel("Year")
    plt.ylabel("Water Balance")
    plt.title("Evolution of Water Balance Over Years")
    plt.show()
```

```python
def run_visualization_pipeline(filepath):
    df = load_and_prepare_data(filepath)

    plot_evaporation(df)
    plot_water_balance(df)
```

```python
run_visualization_pipeline("dataset_bilanhyd_clean_MED.csv")
```
