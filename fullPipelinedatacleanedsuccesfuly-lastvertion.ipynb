```python
import pandas as pd

# 1. Charger le fichier
df = pd.read_csv("dataset-noaa-Copy-toUse.csv", sep=',', skiprows=1)

# 2. Renommer colonnes
df.rename(columns={
    'Date': 'DATE',
    'TAVG (Degrees Fahrenheit)': 'TAVG',
    'TMAX (Degrees Fahrenheit)': 'TMAX',
    'TMIN (Degrees Fahrenheit)': 'TMIN',
    'PRCP (Inches)': 'PRCP',
    'SNOW (Inches)': 'SNOW',
    'SNWD (Inches)': 'SNWD',
}, inplace=True)
```

 CONVERTIR EN NUMÉRIQUE

```python
cols = ["TAVG", "TMAX", "TMIN", "PRCP"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# transforme les valeurs non numériques en NaN
```

 GÉRER LES NaN

```python
df[cols] = df[cols].interpolate(method="linear")
# remplace les valeurs manquantes par interpolation
```

APPLIQUER LES modification sur les donnees preparer

```python
df = df[df["TMAX"] >= df["TMIN"]]
```

```python
df["PRCP"] = df["PRCP"] * 25.4  # inches → mm

df["TAVG"] = (df["TAVG"] - 32) * 5/9
df["TMAX"] = (df["TMAX"] - 32) * 5/9
df["TMIN"] = (df["TMIN"] - 32) * 5/9
```

```python
df["TEMP_RANGE"] = df["TMAX"] - df["TMIN"]

df["TAVG_7D"] = df["TAVG"].rolling(7).mean()
df["PRCP_7D"] = df["PRCP"].rolling(7).mean()

df['dry_index'] = df['PRCP'] / (df['TAVG'] + 1)
```

```python
df = df.round(2)
```

```python
df = df.drop(columns=['SNOW', 'SNWD'])
```

sauvegade

```python
df.to_csv("cleaned_data_version2.csv", index=False)
```

TEMPÉRATURE (moyenne annuelle)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger dataset nettoyé
df = pd.read_csv("cleaned_data_version2.csv")

# Convertir la colonne 'DATE' en format datetime
df["DATE"] = pd.to_datetime(df["DATE"])

# Extraire année
df["YEAR"] = df["DATE"].dt.year

# Filtrer période 1965–2025
df = df[(df["YEAR"] >= 1965) & (df["YEAR"] <= 2025)]


# 1. TEMPÉRATURE (moyenne annuelle)

temp_year = df.groupby("YEAR")["TAVG"].mean()

plt.figure()
plt.plot(temp_year.index, temp_year.values)
plt.title("Évolution annuelle de la température moyenne (1965–2025)")
plt.xlabel("Année")
plt.ylabel("Température (°C)")
plt.grid()
plt.show()
```

PRÉCIPITATION (somme annuelle)

```python
prcp_year = df.groupby("YEAR")["PRCP"].sum()

plt.figure()
plt.plot(prcp_year.index, prcp_year.values)
plt.title("Évolution annuelle des précipitations (1965–2025)")
plt.xlabel("Année")
plt.ylabel("Précipitation (mm)")
plt.grid()
plt.show()
```

 3. SÉCHERESSE

```python
# Définir jours secs (<1 mm)
df["DRY_DAY"] = (df["PRCP"] < 1).astype(int)

# Nombre de jours secs par année
dry_year = df.groupby("YEAR")["DRY_DAY"].sum()

plt.figure()
plt.plot(dry_year.index, dry_year.values)
plt.title("Nombre de jours secs par an (1965–2025)")
plt.xlabel("Année")
plt.ylabel("Jours secs")
plt.grid()
plt.show()
```
