```python
import pandas as pd

# 1. Charger le fichier CSV (⚠️ utiliser read_csv et pas read_excel)
# Le fichier semble être délimité par des virgules et la première ligne est une métadonnée.
df = pd.read_csv("dataset-noaa-Copy-toUse.csv", sep=',', skiprows=1)

# Renommer les colonnes pour simplifier leur utilisation et les faire correspondre au code existant
df.rename(columns={
    'Date': 'DATE', # Renomme 'Date' en 'DATE' pour correspondre à la casse utilisée plus loin
    'TAVG (Degrees Fahrenheit)': 'TAVG',
    'TMAX (Degrees Fahrenheit)': 'TMAX',
    'TMIN (Degrees Fahrenheit)': 'TMIN',
    'PRCP (Inches)': 'PRCP',
    'SNOW (Inches)': 'SNOW',
    'SNWD (Inches)': 'SNWD',
}, inplace=True)

# 2. Vérifier colonnes (après renommage)
print(df.columns)


# 4. Convertir en numérique (une seule fois suffit)
cols = ["TAVG", "TMAX", "TMIN", "PRCP"]
for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# 5. Supprimer lignes totalement vides
df = df.dropna(subset=cols, how="all")

# 6. Cohérence physique
df = df[df["TMAX"] >= df["TMIN"]]

# 7. Conversion unités
df["PRCP"] = df["PRCP"] * 25.4  # inches → mm

df["TAVG"] = (df["TAVG"] - 32) * 5/9
df["TMAX"] = (df["TMAX"] - 32) * 5/9
df["TMIN"] = (df["TMIN"] - 32) * 5/9

# 8. Supprimer valeurs aberrantes (une seule fois suffit)
df = df[(df["TMAX"] < 55) & (df["TMIN"] > -10) & (df["PRCP"] >= 0)]

# 9. Feature Engineering
df["TEMP_RANGE"] = df["TMAX"] - df["TMIN"]
df["TAVG_7D"] = df["TAVG"].rolling(7).mean()
df["PRCP_7D"] = df["PRCP"].rolling(7).mean()

# 10. Arrondir
df = df.round(2)

# 11. Sauvegarde
df.to_csv("cleaned_data.csv", index=False)

print("✅ Nettoyage terminé avec succès")
```
